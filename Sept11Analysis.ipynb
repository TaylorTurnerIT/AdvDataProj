```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import dates as mdates
import seaborn as sns
import warnings as wr
wr.filterwarnings('ignore')
```

```python
# plt.figure(figsize=(15, 6)).autofmt_xdate()
plt.rcParams["figure.figsize"] = [15.00, 6.00]
plt.rcParams["figure.autolayout"] = True

plt.plot()
plt.bar(df['Month'], df['Air RPM (000s)'], color='lightblue', width=30)
plt.title('Count Plot of Air RPM')
plt.xlabel('Month')
plt.ylabel('Air RPMs')
plt.show()
```

```python
# plt.figure(figsize=(15, 6)).autofmt_xdate()
plt.rcParams["figure.figsize"] = [15.00, 6.00]
plt.rcParams["figure.autolayout"] = True

plt.plot()
plt.bar(df['Month'], df['Rail PM'], color='lightblue', width=30)
plt.title('Count Plot of Rail PM')
plt.xlabel('Month')
plt.ylabel('Rail PM')
plt.show()
```
