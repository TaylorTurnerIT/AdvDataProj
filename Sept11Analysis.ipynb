```python
airdf = df['Air RPM (000s)']

plt.figure(figsize=(20, 6)).autofmt_xdate()
fig = plt.plot()
fig.bar(df['Month'], airdf, color='lightblue')
fig.title('Count Plot of Air RPM')
fig.xlabel('Month')
fig.ylabel('Air RPMs')
fig.show()
```
